```python
# importing required packages
import scipy.fftpack as spfft
import scipy.ndimage as spimg
import numpy as np
import matplotlib.pyplot as plt
import imageio

#%% Discrete Cosine Transform 
def dct2(x):
    return spfft.dct(spfft.dct(x.T, norm='ortho', axis=0).T, norm='ortho', axis=0)

def idct2(x):
    return spfft.idct(spfft.idct(x.T, norm='ortho', axis=0).T, norm='ortho', axis=0)

# no of pixels we want to retain
corruption = 0.70;
# path to data
path = 'Data For Assignment/'+str(int(corruption*100))
# declaring a matrix to store the reconstructed image
x_recon70 = np.zeros((100,99,3))
# declaring a matrix to store the corrupted image
x_corr70 = np.zeros((100,99,3))
for i in range(3):
    x_corr70[:,:,i] = imageio.imread(path+'/incomplete'+str(i+1)+'_'+str(int(corruption*100))+'.png',pilmode='L')
    x_corr70[:,:,i] = (x_corr70[:,:,i]-np.min(x_corr70[:,:,i]))/(np.max(x_corr70[:,:,i])-np.min(x_corr70[:,:,i]))
# loading the optimised value s of component 1
s = np.load(path+'/s1_'+str(int(corruption*100))+'.npy')
# reshaping s to 2D matrix of shape (99,100)
s_img = np.reshape(np.array(s),(99,100))
# recovering the image
x_img = idct2(s_img.T)
# scaling the image to get proper values
x_img = (x_img-np.min(x_img))/(np.max(x_img)-np.min(x_img))
# storing component 1 in x_recon
x_recon70[:,:,0] = x_img;

# loading the optimised value s of component 2
s = np.load(path+'/s2_'+str(int(corruption*100))+'.npy');
# reshaping s to 2D matrix of shape (99,100)
s_img = np.reshape(np.array(s),(99,100));
# recovering the image
x_img = idct2(s_img.T);
# scaling the image to get proper values
x_img = (x_img-np.min(x_img))/(np.max(x_img)-np.min(x_img));
# storing component 2 in x_recon
x_recon70[:,:,1] = x_img;

# loading the optimised value s of component 3
s = np.load(path+'/s3_'+str(int(corruption*100))+'.npy');
# reshaping s to 2D matrix of shape (99,100)
s_img = np.reshape(np.array(s),(99,100));
# recovering the image
x_img = idct2(s_img.T);
# scaling the image to get proper values
x_img = (x_img-np.min(x_img))/(np.max(x_img)-np.min(x_img))
# storing component 3 in x_recon
x_recon70[:,:,2] = x_img
```

```python
# reading original image
Xorig = imageio.imread('1.jpg', pilmode='RGB') # read in RGB

# comparing original vs mask vs reconstructed images
plt.figure(figsize=(15,4));
plt.suptitle("Comparison of original vs reconstructed image for corruption = "+str(corruption));
# plotting original image
plt.subplot(131);
plt.imshow(Xorig);
plt.title("Original image");
# plotting original image
plt.subplot(132);
plt.imshow(x_corr70);
plt.title("Corrupted image");
# plotting recovered image
plt.subplot(133);
plt.imshow(x_recon70);
plt.title("Reconstructed image");
```
